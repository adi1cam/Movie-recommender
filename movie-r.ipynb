```python
import numpy as np
import pandas as pd
import ast
```

```python
movie=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movie.head()
```

```python
movie=movie.merge(credits,on='title')
```

```python
movie.head()
```

```python
movie.info()
```

```python
movie=movie[['genres','id','title','keywords','overview','cast','crew']]
```

```python
movie.head()
```

```python
movie.isnull().sum()
```

```python
movie.dropna(inplace=True)
```

movie.duplicated().sum()

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
        
```

```python
movie['genres']=movie['genres'].apply(convert)
```

```python
movie['keywords']=movie['keywords'].apply(convert)
```

```python
def convert3(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
         l.append(i['name'])
         counter+=1
        else:
            break
    
    return l
        
```

```python
movie['cast'].apply(convert3)
```

```python
movie['cast']=movie['cast'].apply(convert3)
```

```python
movie.head()
```

```python
def fetch(obj):
    L= []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
          L.append(i['name'])
          break
    return L
        
```

```python
movie['crew'].apply(fetch)
```

movie.head()

```python
movie['crew']=movie['crew'].apply(fetch)
```

```python
movie.head()
```

```python
movie['overview']=movie['overview'].apply(lambda x:x.split())
```

```python
movie.head()
```

```python
movie['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movie['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movie['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movie['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movie['genres']=movie['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movie['keywords']=movie['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movie['cast']=movie['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movie['crew']=movie['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movie.head()
```

```python
movie['tags'] = movie['overview'] + movie['genres'] + movie['keywords']+movie['cast'] + movie['crew'] 
```

```python
new_df=movie[['id','title','tags']]
```

```python
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
similarity[1]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list= sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
            print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Spectre')
```
